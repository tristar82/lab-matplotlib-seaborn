```python
# Method 1 - matplotlib
# your code here
from numpy.polynomial.polynomial import polyfit
# Sample data
x = df.Age
y = df.Fare

# Fit with polyfit
b, m = polyfit(x, y, 1)

plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')
plt.show()
```
